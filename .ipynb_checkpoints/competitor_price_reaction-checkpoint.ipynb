# Functions

```python
def price_move_classifier(row):
    if row['price_delta_self'] > 0 and row['price_delta_pfj'] > 0:
        return "up_above"
    elif row['price_delta_self'] > 0 and row['price_delta_pfj'] < 0:
        return "up_below"
    elif row['price_delta_self'] > 0 and row['price_delta_pfj'] == 0:
        return "up_equal"
    elif row['price_delta_self'] < 0 and row['price_delta_pfj'] > 0:
        return "down_above"
    elif row['price_delta_self'] < 0 and row['price_delta_pfj'] < 0:
        return "down_below"
    elif row['price_delta_self'] < 0 and row['price_delta_pfj'] == 0:
        return "down_equal"
    else:
        return "unknown"
```

# Data Intake

```python
# reload pickle generated in competitor_price_behavior.ipynb
data_file = '/mnt/hgfs/vmshare/data/augmented_competitor_movements.pkl'
comp_data = pd.read_pickle(data_file)
comp_data['brand'] = comp_data.brand.str.lower()
```

```python
# create a base data set that only includes loves and pfj
base_df = comp_data[(comp_data.brand.str.contains('love|pfj'))].sort_values(by=['unit_id', 'date_id'])
```

```python
# loop over all stores and find the nearest love's and cache the promilesids
stores = base_df.unit_id.unique()
promilesids = list()
for store in tqdm_notebook(stores):
    temp = base_df[(base_df.unit_id == store)]
    if not temp.empty:
        if any("loves" in s for s in temp.brand.unique()):
            loves_min_dist = temp[(temp.brand.str.contains('loves'))].crow_fly_dist.min()
            promilesids.extend(temp[(temp.crow_fly_dist <= loves_min_dist)].promilesid.unique())
```

```python
# filter the original df on these promilesid's
loves_df = base_df[(base_df.promilesid.isin(promilesids))]

# sort data by unit and date
loves_df = loves_df.sort_values(['unit_id', 'date_id'])

# calc a reaction time to prev price move
loves_df['reaction_time'] = loves_df.groupby('unit_id')['date_id'].diff().dt.total_seconds()/3600.

# encode the move as being to self or competitor
loves_df['reaction_type'] = loves_df.groupby('unit_id')['brand'].apply(lambda x: x.shift() == x)
loves_df.loc[loves_df.reaction_type == True, 'reaction_type'] = 'self'
loves_df.loc[loves_df.reaction_type == False, 'reaction_type'] = 'comp'

# encode the move as being before or after a critical date
crit_date = dt.datetime(2017,7,1)
today = loves_df.date_id.max()
diff = (today - crit_date).days
start_date = (crit_date - dt.timedelta(days=diff)).date()
loves_df = loves_df[(loves_df.date_id >= start_date)]
loves_df['era'] = 'none'
loves_df.loc[loves_df.date_id < crit_date, 'era'] = 'pre'
loves_df.loc[loves_df.date_id >= crit_date, 'era'] = 'post'

# calc the price move
loves_df['price_delta_pfj'] = loves_df.groupby('unit_id')['dslprice'].diff()

# calc the price move type
loves_df['move_type'] = loves_df.apply(price_move_classifier, axis=1)

# bulk encode the move type
loves_df['move_direction'] = 'none'
loves_df.loc[loves_df.price_delta_self < 0, 'move_direction'] = 'down'
loves_df.loc[loves_df.price_delta_self > 0, 'move_direction'] = 'up'

# remove nan's
#loves_df = loves_df.dropna()
```

```python
# filter the df for this data and to include love's only
rxn_cut = 24*7
criterion = (loves_df.brand.str.contains('loves')) & \
            (loves_df.reaction_type == 'comp') & \
            (loves_df.reaction_time <= rxn_cut) & \
            (loves_df.move_type != 'unknown')
loves_comp_rxn = loves_df[criterion]
```

# Data Exploration

```python
# what % of moves are supported before critical date?
temp = loves_df[(loves_df.reaction_time > 0) & (loves_df.date_id < crit_date)]
temp['supported'] = temp.groupby('unit_id')['brand'].apply(lambda x: x.shift(-1) != x)
g = temp[(temp.brand == 'pfj')].groupby('supported')
g.date_id.count()/g.date_id.count().sum()
```

```python
# what % of moves are supported after critical date?
temp = loves_df[(loves_df.reaction_time > 0) & (loves_df.date_id >= crit_date)]
temp['supported'] = temp.groupby('unit_id')['brand'].apply(lambda x: x.shift(-1) != x)
g = temp[(temp.brand == 'pfj')].groupby('supported')
g.date_id.count()/g.date_id.count().sum()
```

```python
# get breakdown of move types
moves = sorted(loves_comp_rxn.move_type.unique().tolist())
moves.append('|'.join(moves))
moves = moves[::-1]
f, axs = plt.subplots(len(moves))

for i, move in enumerate(moves):
    
    # summarize
    num_move = len(loves_comp_rxn[(loves_comp_rxn.move_type.str.contains(move))])
    perc_total = 100 * num_move / len(loves_comp_rxn)
    print('Number of {} moves: {}'.format(move, num_move))
    print('%Total of {} moves: {:.1f}%'.format(move, perc_total))

    # density plot of all down
    a = loves_comp_rxn[(loves_comp_rxn.era == 'pre') & (loves_comp_rxn.move_type.str.contains(move))]
    g = sns.kdeplot(a.reaction_time, shade=True, color='r', ax=axs[i])
    b = loves_comp_rxn[(loves_comp_rxn.era == 'post') & (loves_comp_rxn.move_type.str.contains(move))]
    g = sns.kdeplot(b.reaction_time, shade=True, color='b', ax=axs[i])
    g.set_xlim(0, rxn_cut)
    g.set_title(move)
    #g.set_ylim(0, 0.06)
    
    # t-test
    tresults = sp.stats.ttest_ind(a.reaction_time.values, b.reaction_time.values)
    print('{:12}: {:.2f}'.format('T-Statistic', tresults.statistic))
    print('{:12}: {:.2e}'.format('p-value', tresults.pvalue))
    if tresults.pvalue < 0.05:
        outcome = 'REJECT'
    else:
        outcome = 'ACCEPT'
    print('{} the null hypothesis that reaction time has not changed\n\n'.format(outcome))
    
fig = plt.gcf()
fig.set_size_inches(15, 2*len(axs))
```
