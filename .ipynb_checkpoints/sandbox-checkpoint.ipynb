```python
def calculate_nan_values(orig_df):
    df = orig_df.copy()
    df = df.isna().sum(axis=0).reset_index()
    df.columns = ['column_name', 'na_count']
    df['na_ratio'] = df['na_count'] / orig_df.shape[0]
    df = df.loc[df['na_ratio'] > 0]
    return df

def calculate_null_values(orig_df):
    df = orig_df.copy()
    df = df.isnull().sum(axis=0).reset_index()
    df.columns = ['column_name', 'null_count']
    df['null_ratio'] = df['null_count'] / orig_df.shape[0]
    df = df.loc[df['null_ratio'] > 0]
    return df

def value_limits(df, field, upper_bound=99, lower_bound=1):
    '''
    Find the upper and lower limits of data based on quantiles; for quick outlier removal
    '''
    upper_limit = np.percentile(df[field].values, upper_bound)
    lower_limit = np.percentile(df[field].values, lower_bound)
    return upper_limit, lower_limit

def visualize_correlations(df, fields):

    # keep a copy
    df = df.copy()

    # correlation coefficient of each column
    for field in fields:
        x_cols = [col for col in df.columns
                  if col != field and df[col].dtype=='float64']
        labels = []
        values = []
        for col in x_cols:
            labels.append(col)
            values.append(
                np.corrcoef(df[col].values, df[field].values)[0, 1])

        # create dataframe for corr coeffs
        corr_df = pd.DataFrame({'col_labels': labels, 'corr_values': values})
        corr_df = corr_df.sort_values(by='corr_values')

        ind = np.arange(len(labels))
        fig, ax = plt.subplots(figsize=(12, 4))
        rects = ax.barh(ind, np.array(corr_df.corr_values.values), color='y')
        ax.set_yticks(ind)
        ax.set_yticklabels(corr_df.col_labels.values, rotation='horizontal')
        ax.set_xlabel('correlation coefficient')
        ax.set_title('Correlations with {}'.format(field))
        ax.set_xlim(-1, 1)
        plt.show()

def visualize_values(df, fields):
    
    # keep a copy
    df = df.copy()

    # iterate over fields
    for field in fields:

        # get limits of data
        ulimit, llimit = value_limits(df, field)
        print('Upper limit: {:.3f}'.format(ulimit))
        print('Lower limit: {:.3f}'.format(llimit))

        # plot the ordered values for field
        plt.figure(figsize=(12, 8))
        plt.scatter(range(df.shape[0]), np.sort(df[field].values))
        plt.xlabel('index', fontsize=12)
        plt.ylabel(field, fontsize=12)
        plt.show()

        # plot histo of values
        sns.distplot(df[field].values, bins=50, kde=True, norm_hist=True)
        plt.xlabel(field, fontsize=12)
        plt.show()
```

```python
def weekday_of_month(date):
    '''
    Calculate the number of times the current weekday has appeared in the month
    '''
    miter = calendar.Calendar().itermonthdays2(date.year, date.month)
    wd = date.weekday()
    d = date.day
    wday7 = [1 for x in miter if x[0] != 0 and x[1] == wd and x[0] < d]
    return len(wday7) + 1
```

```python
def is_weekend(date):
    '''
    Calculate the number of times the current weekday has appeared in the month
    '''
    wd = date.weekday()
    if wd in [5, 6]:
        return True
    else:
        return False
```

```python
def day_of_quarter(date):
    '''
    Calculate the days into the present quarter
    '''
    qstart = dt.datetime(date.year, 3 * date.quarter - 2, 1)
    qday = date.toordinal() - qstart.toordinal() + 1
    return qday
```

```python
def days_until_taxes(date):
    '''
    Calculate the days into the present tax quarter
    '''
    tax_qs = [
        dt.datetime(date.year, 4, 15).toordinal(),
        dt.datetime(date.year, 6, 15).toordinal(),
        dt.datetime(date.year, 9, 15).toordinal(),
        dt.datetime(date.year, 12, 15).toordinal(),
        dt.datetime(date.year + 1, 4, 15).toordinal()
    ]
    date = date.toordinal()
    if date <= tax_qs[0]:
        tday = tax_qs[0] - date
    elif date <= tax_qs[1]:
        tday = tax_qs[1] - date
    elif date <= tax_qs[2]:
        tday = tax_qs[2] - date
    elif date <= tax_qs[3]:
        tday = tax_qs[3] - date
    else:
        tday = tax_qs[4] - date
    return tday
```

```python
class TruckerCalendar(pdh.AbstractHolidayCalendar):

    # create a class that inherits the generic fed holiday class
    rules = pdh.get_calendar('USFederalHolidayCalendar').rules

    # add custom rules or dates for holidays
    rules += [
        pdh.Holiday('Xmas Eve', month=12, day=24),
        pdh.Holiday('Xmas', month=12, day=25),
        pdh.Holiday('New Years', month=1, day=1)
    ]
```

```python
def create_holiday_calendar(start_date, end_date):
    """
    Create a holiday dataframe unique to our business

    :param start_date: the date to start calendar
    :param end_date: the date to end calendar
    :return: dataframe of holidays
    """

    calendar = TruckerCalendar()
    hdays = calendar.holidays(start_date, end_date)
    return hdays
```

```python
def create_uuid(df, columns, uuid_name='uuid'):
    '''
    Create a unique identifier for a dataframe based on some columns
    '''
    
    # ensure cleanliness
    df = df.copy()
    
    # add a unique identifier for each row
    df[uuid_name] = ''
    for column in columns:
        df[uuid_name] += df[column].astype(str)
    df[uuid_name] = (
        df.uuid).apply(lambda x: hashlib.md5(x.encode('utf-8')).hexdigest())
    ids = len(df[uuid_name])
    uids = len(df[uuid_name].unique())
    puni = 100 * uids / ids
    
    # report on uniqueness
    print('Created {} hashes, {} are unique ({}%)'.format(ids, uids, puni))
    
    return df
```

```python
def create_time_features(df, date_field):
    '''
    Create numerous time features for a dataframe with a date column
    '''
    
    # ensure cleanliness
    df = df.copy()
    
    # get holidays
    start_date = df.date_id.min()
    end_date = df.date_id.max()
    holidays = create_holiday_calendar(start_date, end_date)
    
    # calculate a bunch of time features
    df[date_field] = df[date_field].apply(lambda x: x.to_pydatetime())
    df['year'] = df[date_field].apply(lambda x: x.year)
    df['quarter'] = df[date_field].apply(lambda x: x.quarter)
    df['month'] = df[date_field].apply(lambda x: x.month)
    df['day'] = df[date_field].apply(lambda x: x.day)
    df['hour'] = df[date_field].apply(lambda x: x.hour)
    df['minute'] = df[date_field].apply(lambda x: x.minute)
    df['second'] = df[date_field].apply(lambda x: x.second)
    df['weekday'] = df[date_field].apply(lambda x: x.weekday())
    df['monthday'] = df[date_field].apply(lambda x: weekday_of_month(x))
    df['quarterday'] = df[date_field].apply(lambda x: day_of_quarter(x))
    df['yearday'] = df[date_field].apply(lambda x: x.toordinal() - dt.date(x.year, 1, 1).toordinal() + 1)
    df['monthweek'] = df[date_field].apply(lambda x: (x.day - 1) // 7 + 1)
    df['quarterweek'] = df.quarterday.apply(lambda x: (x - 1) // 7 + 1)
    df['yearweek'] = df[date_field].apply(lambda x: x.isocalendar()[1])
    df['weekend'] = df[date_field].apply(lambda x: is_weekend(x))
    df['holiday'] = df[date_field].isin(holidays)
    df['taxday'] = df[date_field].apply(lambda x: days_until_taxes(pd.to_datetime(x)))
    
    return df
```

```python
def create_rolling_features(df, sort_by, group_by, modify_cols, windows):
    '''
    Create roling sum and mean features from input columns
    '''

    # ensure cleanliness
    df = df.copy()
    
    # rolling sums
    for c in tqdm_notebook(modify_cols):
        for window in windows:
            new_col = 'rt_{}d_{}'.format(window, c)
            df[new_col] = df.sort_values(by=sort_by).groupby(group_by).rolling(
                window=window)[c].sum().round().values
            new_col = 'ra_{}d_{}'.format(window, c)
            df[new_col] = df.sort_values(by=sort_by).groupby(group_by).rolling(
                window=window)[c].mean().round().values         

    # fillna with zeroes
    df = df.fillna(0)
    
    return df
```

## Trees

```python
# create a train df
train_df = merged_df.copy().reset_index()

# set target var
target = 'ccc_gallons'
train_y = train_df[target].values

# remove unwanted features
drop_cols = [
    'date_id_v', 'unit_id_v', 'uuid', 'total_cost', 'total_gallons_v',
    'retail_gallons', 'ccc_perc_total', 'retail_perc_total', target
]
train_df = train_df.drop(drop_cols, axis=1)
feat_names = train_df.columns.values
```

## Autoencoder

```python
# train test split iterator for cv purposes
X = merged_df.copy()
```

```python
# isolate the features and target
target = 'retail_price'
y = X[target]
X.drop(target, axis=1, inplace=True)
#X = X[X.columns.drop(list(X.filter(regex='_gallons|_perc')))]

# create a train set and true test set
test_index = math.floor(0.25 * len(X))
X_test = X.iloc[-test_index:]
y_test = y.iloc[-test_index:].values
X = X.drop(X.index[-test_index:])
y = y.drop(y.index[-test_index:]).values
```

```python
# build the model
model = H2OAutoEncoderEstimator(
    ignored_columns=['date_id'],
    activation="tanh_with_dropout",
    hidden=[50],
    l1=1e-5,
    l2=1e-5,
    epochs=100)

# k-folding training
tscv = model_selection.TimeSeriesSplit(n_splits=5)
for train_index, cv_index in tscv.split(X):

    # train/cv split
    X_train, X_cv = h2o.H2OFrame(X.iloc[train_index]), h2o.H2OFrame(
        X.iloc[cv_index])
    y_train, y_cv = h2o.H2OFrame(y[train_index]), h2o.H2OFrame(y[cv_index])

    # factorize the binary cols
    dfs = [X_train, X_cv]
    factor_cols = ['weekend', 'holiday']
    for df in dfs:
        for c in factor_cols:
            df[c] = df[c].asfactor()

    # train the model
    model.train(x=X_train.names, training_frame=X_train)
```

```python
# convert to H2O's df format
automl_df_h2o = h2o.H2OFrame(automl_df)
test_h2o = h2o.H2OFrame(automl_df_test)
train_h2o, cv_h2o = automl_df_h2o.split_frame(ratios=[0.8])
```

```python
# error calculations
err_df = automl_df_test[(automl_df_test[target] > 0)][[target, 'pred']]
err_df['pred_err'] = err_df[target] - err_df.pred
err_df['pred_errpct'] = err_df.pred_err / err_df[target]
errors = []
errors.append(('Mean Err', err_df.pred_err.mean()))
errors.append(('Mean Err%', err_df.pred_errpct.mean()))
errors.append(('RMSE', ((err_df.pred_err)**2).mean()**.5))
errors.append(('Mean AbsErr', err_df.pred_err.abs().mean()))
errors.append(('Mean AbsErr%', err_df.pred_errpct.abs().mean()))
```
